## Begin with cleaning the Human Development Index dataset

- HDI Category -> Country is placed within one of four categories based on HDI score

1. Very High Development Country
2. High Development Country
3. Medium Development Country
4. Low Development Country

- Education Index is a formula based on the two schooling sub-dimensions.


```python
# Removed unnamed columns - these were used as 'footnote' letters
df_extras_removed = df_extras_removed.drop(['Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9','Unnamed: 11','Unnamed: 13','Unnamed: 15'],axis=1)
df_extras_removed.head()
```

```python
# Rename current columns
df_extras_removed.rename(columns={'Unnamed: 0': 'HDI Rank (2021)', 'Unnamed: 1': 'Countries', 'HDI rank': 'HDI rank (2020)','Gross national income (GNI) per capita': 'Gross national income (GNI) per capita (2017 PPP $)'},inplace=True)
df_extras_removed = df_extras_removed.reset_index(drop=True)
df_extras_removed.head()
```

```python
# Creating an excel for all countries - HDI only so far
cleaned_all_HDI = df_extras_removed.iloc[:194,:]
print(cleaned_all_HDI.shape)
# Remove all rows with null values - mostly there are rows under 'country' where it labels "High/Medium/Low development HDI" 
cleaned_all_HDI.dropna(axis=0, how='any',inplace=True)
print(cleaned_all_HDI.shape)
```

```python
# Create new column for HDI 'level' (1 = Very high, 2 = high, 3 = medium, 4 = low)
HDI_development_category = []
counter = 1
for i in cleaned_all_HDI["Countries"]:
    if "Albania" in i:
        counter += 1
    elif "Philippines" in i:
        counter += 1
    elif "Tanzania (United Republic of)" in i:
        counter += 1
    HDI_development_category.append(counter)
    
cleaned_all_HDI.loc[:, "HDI Category"] = HDI_development_category
cleaned_all_HDI.head()
```

```python
# Calculating education index
cleaned_all_HDI["Education Index"] = ((cleaned_all_HDI["Expected years of schooling"] / 18) + (cleaned_all_HDI["Mean years of schooling"] / 15)) / 2
cleaned_all_HDI.head()
```

## Web Scraping Numbeo's dataset - City level data

Data obtained from <https://www.numbeo.com/cost-of-living/>

```python
def LPP_city():
    url = "https://www.numbeo.com/cost-of-living/rankings.jsp"
    tables = pd.read_html(url)
    city_table = tables[1]
    del city_table['Rank']
    del city_table['Cost of Living Index']
    del city_table['Rent Index']
    del city_table['Cost of Living Plus Rent Index']
    del city_table['Groceries Index']
    del city_table['Restaurant Price Index']
    return city_table

def crime_city():
    url = "https://www.numbeo.com/crime/rankings.jsp"
    tables = pd.read_html(url)
    city_table = tables[1]
    del city_table['Rank']
    del city_table['Safety Index']
    return city_table

def healthcare_city():
    url = "https://www.numbeo.com/health-care/rankings.jsp"
    tables = pd.read_html(url)
    city_table = tables[1]
    del city_table['Rank']
    del city_table['Health CareExp. Index']
    return city_table
```

```python
# Using the functions to clean via Pandas
LPP_city_table = LPP_city()
crime_city_table = crime_city()
healthcare_city_table = healthcare_city()
LPP_city_table
```

```python
LPP_city_table.columns = ['city','LPP_index']
crime_city_table.columns = ['city','crime_index']
healthcare_city_table.columns = ['city','healthcare_index']


merge2_df = pd.merge(LPP_city_table, crime_city_table, how='left', on='city')
merge3_df = pd.merge(merge2_df, healthcare_city_table, how='left',on='city')
merge3_df.rename(columns={'city':'city_country'},inplace=True)
merge3_df.head()
```

### Used difflib (used for fuzzy matching/approximate string matching) - to search for match city variables with latitude and longitude data from another dataset (given spelling differences).

City dataset courtesy of <https://simplemaps.com/data/world-cities>

```python
world_cities_df = pd.read_csv(os.path.join("dataset","worldcities.csv"),header=0)
world_cities_df.head()
```

```python
world_cities_df.drop(columns=['city','iso2','admin_name','population','id'], inplace=True)
world_cities_df['city_country'] = world_cities_df['city_ascii'] + ', ' + world_cities_df['country']
world_cities_df.head()
```

```python
import difflib


# Function to find the top fuzzy match
def find_top_match(city_country, candidates, cutoff=0.6):
    matches = difflib.get_close_matches(city_country, candidates, n=1, cutoff=cutoff)
    return matches[0] if matches else None

# Remove duplicate rows based on the 'city_country' column
world_cities_df_unique = world_cities_df.drop_duplicates(subset='city_country', keep='first')

# Create a mapping of city_country to lat and lon
city_country_to_lat_lon = world_cities_df_unique.set_index('city_country')[['lat', 'lng', 'iso3']].to_dict(orient='index')

# Find top matches for each city_country in merge3_df
merge3_df['top_match'] = merge3_df['city_country'].apply(lambda x: find_top_match(x, city_country_to_lat_lon.keys()))

# Extract lat and lon for matched city_country and store them in merge3_df
merge3_df['lat'] = merge3_df['top_match'].apply(lambda x: city_country_to_lat_lon.get(x, {}).get('lat'))
merge3_df['lon'] = merge3_df['top_match'].apply(lambda x: city_country_to_lat_lon.get(x, {}).get('lng'))
merge3_df['alpha-3'] = merge3_df['top_match'].apply(lambda x: city_country_to_lat_lon.get(x, {}).get('iso3'))
```
