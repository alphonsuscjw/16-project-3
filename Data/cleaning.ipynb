## Begin with cleaning the Human Development Index dataset

- HDI Category -> Country is placed within one of four categories based on HDI score

1. Very High Development Country
2. High Development Country
3. Medium Development Country
4. Low Development Country

- Education Index is a formula based on the two schooling sub-dimensions.


```python
# Removed unnamed columns - these were used as 'footnote' letters
df_extras_removed = df_extras_removed.drop(['Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9','Unnamed: 11','Unnamed: 13','Unnamed: 15'],axis=1)
df_extras_removed.head()
```

```python
# Rename current columns
df_extras_removed.rename(columns={'Unnamed: 0': 'HDI Rank (2021)', 'Unnamed: 1': 'Countries', 'HDI rank': 'HDI rank (2020)','Gross national income (GNI) per capita': 'Gross national income (GNI) per capita (2017 PPP $)'},inplace=True)
df_extras_removed = df_extras_removed.reset_index(drop=True)
df_extras_removed.head()
```

```python
# Creating an excel for all countries - HDI only so far
cleaned_all_HDI = df_extras_removed.iloc[:194,:]
print(cleaned_all_HDI.shape)
# Remove all rows with null values - mostly there are rows under 'country' where it labels "High/Medium/Low development HDI" 
cleaned_all_HDI.dropna(axis=0, how='any',inplace=True)
print(cleaned_all_HDI.shape)
```

```python
# Create new column for HDI 'level' (1 = Very high, 2 = high, 3 = medium, 4 = low)
HDI_development_category = []
counter = 1
for i in cleaned_all_HDI["Countries"]:
    if "Albania" in i:
        counter += 1
    elif "Philippines" in i:
        counter += 1
    elif "Tanzania (United Republic of)" in i:
        counter += 1
    HDI_development_category.append(counter)
    
cleaned_all_HDI.loc[:, "HDI Category"] = HDI_development_category
cleaned_all_HDI.head()
```

```python
# Calculating education index
cleaned_all_HDI["Education Index"] = ((cleaned_all_HDI["Expected years of schooling"] / 18) + (cleaned_all_HDI["Mean years of schooling"] / 15)) / 2
cleaned_all_HDI.head()
```

# Grabbing country codes (alpha-3)/continent/regions

Data courtesy of <https://github.com/lukes/ISO-3166-Countries-with-Regional-Codes>

```python
df2 = pd.read_csv("all.csv", header=0)
cleaned_df2 = df2.loc[:,['name', 'alpha-3','region','sub-region']]
## Fixing notable countries' spelling
cleaned_df2.replace(['United States of America','United Kingdom of Great Britain and Northern Ireland', 'Korea, Republic of'], ['United States','United Kingdom','Korea (Republic of)'],inplace=True)
cleaned_df2.head()
```

```python
# Merge them together with original dataframe
cleaned_all_HDI2 = pd.merge(cleaned_all_HDI,cleaned_df2,how="inner",left_on="Countries",right_on="name")
cleaned_all_HDI2.drop(columns=['name'],axis=1,inplace=True)
cleaned_all_HDI2.head()


```

# Grabbing Latitude and longitude

Data courtesy of <https://gist.github.com/tadast/8827699>

```python
df3 = pd.read_csv("countries_codes_and_coordinates.csv", header=0)
cleaned_df3 = df3.loc[:,['Country', 'Alpha-3 code','Latitude (average)','Longitude (average)']]

cleaned_df3 = cleaned_df3[['Country','Alpha-3 code','Latitude (average)','Longitude (average)']].apply(lambda s:s.str.replace('"', ''))
cleaned_df3 = cleaned_df3[['Country','Alpha-3 code','Latitude (average)','Longitude (average)']].apply(lambda s:s.str.replace(' ', ''))
cleaned_df3 = cleaned_df3.rename(columns={'Alpha-3 code':"alpha-3", "Latitude (average)": "Lat","Longitude (average)": "Lon" })

cleaned_df3
```

```python
# Merging with original dataframe
cleaned_all_HDI3 = pd.merge(cleaned_all_HDI2,cleaned_df3,on='alpha-3')
cleaned_all_HDI3.drop(columns=['Country'],axis=1,inplace=True)
cleaned_all_HDI3.head()
```

```python
cleaned_all_HDI3.to_csv("All_HDI2.csv",index=False)
```

## Web Scraping Numbeo's dataset - Local Purchasing Power (LPP), Health Care Index, Crime Index

### This section focuses on country-level data.

Data obtained from <https://www.numbeo.com/cost-of-living/>

```python
# Scraping Numbeo's data for Cost of Living (Local Purchasing Power Index), Health Care Index, Crime Index - Countries
import lxml

# Loocal Purchasing Power Info
def LPP_country():
    url = "https://www.numbeo.com/cost-of-living/rankings_by_country.jsp"
    tables = pd.read_html(url)
    country_table = tables[1]
    del country_table['Rank']
    del country_table['Cost of Living Index']
    del country_table['Rent Index']
    del country_table['Cost of Living Plus Rent Index']
    del country_table['Groceries Index']
    del country_table['Restaurant Price Index']
    return country_table

def crime_country():
    url = "https://www.numbeo.com/crime/rankings_by_country.jsp"
    tables = pd.read_html(url)
    country_table = tables[1]
    del country_table['Rank']
    del country_table['Safety Index']
    return country_table

def healthcare_country():
    url = "https://www.numbeo.com/health-care/rankings_by_country.jsp"
    tables = pd.read_html(url)
    country_table = tables[1]
    del country_table['Rank']
    del country_table['Health CareExp. Index']
    return country_table
    
```

```python
# Using the functions to clean via Pandas
LPP_country_table = LPP_country()
crime_country_table = crime_country()
healthcare_country_table = healthcare_country()

### Extra Code for if you want it vertically stacked
# numbeo_country_df = pd.concat([LPP_country_table, crime_country_table, healthcare_country_table], ignore_index=True, axis=1)
# numbeo_country_df.columns = ['Country', "LPP_index", 'Country2','Crime_index','Country3','Healthcare_index'] 
# numbeo_country_df.head()
```

```python
# Column name changes + merging into one (crime table has the most data, healthcare has the least)
LPP_country_table.columns = ['country','LPP_index']
crime_country_table.columns = ['country','crime_index']
healthcare_country_table.columns = ['country','healthcare_index']

test2 = pd.merge(crime_country_table, LPP_country_table, how='left', on='country')
test3 = pd.merge(test2, healthcare_country_table, how='left',on='country')
test3.head()
```

```python
numbeo_country_df = pd.merge(test3, cleaned_df2, how='left',left_on='country',right_on='name')
numbeo_country_df.drop(columns=['region','sub-region','name'],axis=1, inplace=True)
numbeo_country_df.to_excel('numbeo_country.xlsx',index=False)
```
